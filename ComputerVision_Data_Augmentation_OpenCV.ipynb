### Import required packages

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

### Load the base image, convert to RGB and display the image

```python
img=cv2.imread('/content/veena_lady.png')
img_resize=cv2.resize(img, (225,225))
rgb_img_resize = cv2.cvtColor(img_resize, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(6, 6))
plt.imshow(rgb_img_resize)
plt.title("Original Image")
plt.axis('off')
plt.show()
```

### Generate a few images with some variations on the base image

```python
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
i = 0
augmented_images = []
for batch in datagen.flow(rgb_img_resize, batch_size=1):
    augmented_images.append(batch[0].astype('uint8'))
    i += 1
    if i % 8 == 0:
        break
```

### Show the new images

```python
fig, axes = plt.subplots(2, 4, figsize=(10, 5))
axes = axes.flatten()
for img, ax in zip(augmented_images, axes):
    ax.imshow(img)
    ax.axis('off')
plt.suptitle("Augmented Images")
plt.show()
```
