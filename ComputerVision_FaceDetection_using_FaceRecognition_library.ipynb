### Import required modules

````python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import face_recognition
````

### Load the image, convert to gray and display the image

````python
img=face_recognition.load_image_file('/content/virushka.jpeg')
img_resize=cv2.resize(img, (255,255))
````

````python
plt.figure(figsize=(5, 10))
plt.imshow(img_resize)
plt.axis('off')
plt.show()
````

###Detect the face coordinates and draw bounding box on the face

````python
faces=face_recognition.face_locations(img_resize)
for face in faces:
    cv2.rectangle(img_resize, (face[3], face[0]),(face[1], face[2]), (255,0,255), 2)
````

### Show the image with the bounding box on the faces

````python
plt.figure(figsize=(5, 10))
plt.imshow(img_resize)
plt.title("Face with bounding box")
plt.axis('off')
plt.show()
````
