### Import required modules

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

### Load the image, convert to gray and display the image

```python
img=cv2.imread('/content/virushka.jpeg')
img_resize=cv2.resize(img, (255,255))
gray_img_resize = cv2.cvtColor(img_resize, cv2.COLOR_BGR2GRAY)
gray_img_resize.shape
```

```python
plt.figure(figsize=(10, 20))
plt.imshow(gray_img_resize)
plt.title("Gray Image")
plt.axis('off')
plt.show()
```

### Load the classifier, detect the face coordinates and draw bounding box on the face

```python
face_classifier = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
face = face_classifier.detectMultiScale(
    gray_img_resize, scaleFactor=1.1, minNeighbors=5, minSize=(40, 40)
)
for (x, y, w, h) in face:
    cv2.rectangle(img_resize, (x, y), (x + w, y + h), (0, 255, 0), 4)
```

### Show the image with the bounding box on the faces

```python
img_rgb=cv2.cvtColor(img_resize, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(6, 6))
plt.imshow(img_rgb)
plt.title("Face with bounding box")
plt.axis('off')
plt.show()
```
