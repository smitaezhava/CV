#### Import required modules

```python
import face_recognition
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

#### Load the input images and display them

```python
img1=face_recognition.load_image_file('C:/Users/Lenovo-pc/Downloads/virat2.jpg')
img1_resize=cv2.resize(img1, (255,255))
img2=face_recognition.load_image_file('C:/Users/Lenovo-pc/Downloads/virushka.jpeg')
img2_resize=cv2.resize(img2, (255,255))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(5, 10))
axes[0].imshow(img1_resize)
axes[0].axis('off')
axes[1].imshow(img2_resize)
axes[1].axis('off')
axes[0].title.set_text('Virushka')
axes[1].title.set_text('Virat')
plt.show()
```

#### Detect the faces and store the face encodings and locations

```python
faces1_encoding=face_recognition.face_encodings(img1_resize)
faces2_encoding=face_recognition.face_encodings(img2_resize)

faces1=face_recognition.face_locations(img1_resize)
faces2=face_recognition.face_locations(img2_resize)
```

#### Compare the faces in both the images and draw bounding boxes on the faces which match

```python
for ctr1 in range(len(faces1_encoding)):
    for ctr2 in range(len(faces2_encoding)):
        is_same=False
        is_same=face_recognition.compare_faces([faces1_encoding[ctr1]], faces2_encoding[ctr2])
        if str(is_same) == '[True]':
            cv2.rectangle(img1_resize, (faces1[ctr1][3], faces1[ctr1][0]),(faces1[ctr1][1], faces1[ctr1][2]), (255,0,255), 2)
            cv2.rectangle(img2_resize, (faces2[ctr2][3], faces2[ctr2][0]),(faces2[ctr2][1], faces2[ctr2][2]), (255,0,255), 2)
            fig, axes = plt.subplots(1, 2, figsize=(5, 10))
            axes[0].imshow(img1_resize)
            axes[0].axis('off')
            axes[1].imshow(img2_resize)
            axes[1].axis('off')
            axes[0].title.set_text('Virat1')
            axes[1].title.set_text('Virat2')
            plt.show()
```
